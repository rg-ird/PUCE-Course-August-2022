# Taller: entender el formato fast5

Para entender mejor el asunto, preparamos un pequeño conjunto de datos de un genoma de Bacillus subtilis, Escherichia coli o de drosophila que fue secuenciado con el nanoporo #R10. La salida de cualquier secuenciador Oxford Nanopore se almacena en formato fast5 (el formato bruto de nanopore) (https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=&ved=2ahUKEwjl78ag6YL5AhWtQTABHWM8Cu8QFnoECAIQAw&url=https%3A%2F%2Fwww.biorxiv.org%2Fcontent%2Fbiorxiv%2Fearly%2F2021%2F06%2F30%2F2021.06.29.450255%2FDC2%2Fembed%2Fmedia-2.pdf%3Fdownload%3Dtrue&usg=AOvVaw1Qc6P-3YAJRPhjk-PVh89o),  que es una especialización del formato de archivo [hdf5](https://www.hdfgroup.org/solutions/hdf5/). Es un formato de datos jerárquico, que se parece a un sistema de archivos, utilizado para almacenar y organizar grandes cantidades de datos. El formato es compatible con una amplia gama de lenguajes de programación, y la API de Python está incorporada en el módulo `h5py`.

```python
import h5py

fast5 = h5py.File('data/bacillus_subtilis/bs_raw.fast5', mode='r')
fast5.name
```

```python
list(fast5.keys())
```

```python
len(fast5.keys())
```

El objeto que cargamos del archivo fast5 es tratado como un simple diccionario de Python. Para obtener una lectura, simplemente ponemos el nombre de la lectura deseada dentro de los paréntesis `[]`. También podemos ver todo el árbol de objetos con la función miembro visit. Además, cada objeto tiene metadatos en su variable miembro `attrs`.

Usually, the basecalling is accompanied with quality control, classifying reads into pass/faill classes. Filtering is done with regards to the reads average Phred quality score. For each base, the score is calculated with: $Q = -10log_{10}{P}$, where $P$ is the base-calling error probability. For example, a Phred score of $50$ equals the base call accuracy of $99.999\%$. Guppy has options `--qscore_filtering` and `--min_qscore`, first enabling filtering and the second setting the threshold (which is 7 by default). To store all found data into a single fastq file use option `-q 0`. Let us now try to basecall the whole dataset.

```python
!ont-guppy-cpu/bin/guppy_basecaller \
    -c ont-guppy-cpu/data/dna_r10_450bps_fast.cfg \
    -i data/bacillus_subtilis/ \
    -s guppy_all \
    -q 0 \
    --qscore_filtering \
    --min_qscore 7
```

To see the the length distribution of the basecalled read set we can simply draw a histogram as below. In addition, it would be nice to calculate the average base accuracy given a reference genome. Lets do that for one read that passed the filtering.

```python
from Bio import SeqIO

with open('guppy_all/pass/fastq_runid_aea170ff77e7e76bd42c58335c528e15f49ba94c_0_0.fastq') as handle:
    reads = []
    for read in SeqIO.parse(handle, 'fastq'):
        reads.append(len(read.seq))
    pyplot.hist(reads)
```

Lets pick read `04f26824-995e-49df-b661-7bf4efcc8ce9` for example. First we need to find where in the reference did it originate. This is done with tools called mappers which will be explained in the next session. For now we can assume we know the positions are $[2158952, 2182735]$. We need to extract the read and the portion of the reference into new files. Afterwards we will align them with edit distance to calculate the accuracy.

```python
with open('data/bacillus_subtilis/bs_ref.fasta') as handle:
    for reference in SeqIO.parse(handle, 'fasta'):
        reference.seq = reference.seq[2158952:2182735]
        SeqIO.write(reference, 'ref_substr_1.fasta', 'fasta')

with open('guppy_all/pass/fastq_runid_aea170ff77e7e76bd42c58335c528e15f49ba94c_0_0.fastq') as handle:
    for read in SeqIO.parse(handle, 'fastq'):
        if read.name == '04f26824-995e-49df-b661-7bf4efcc8ce9':
            SeqIO.write(read, 'read_1.fasta', 'fasta')
            break;
```

```python
!cat ref_substr_1.fasta
```

```python
!cat read_1.fasta
```

To calculate the edit distance, which is the number of modifications to get one string from the other, we will be using `edlib`, probably the fastest implementation of the algorithm as it uses the bit-vector approach by Myers.

```python
!edlib-aligner
```

```python
!edlib-aligner \
    read_1.fasta \
    ref_substr_1.fasta
```

The edit distance is `3457`, given the sequence length `23050` we can calculate the approximate read accuracy by dividing those two numbers and get `0,85002`. We can also see the full alignment between the read and reference by passing option `-p`.

```python
!edlib-aligner \
    -p \
    read_1.fasta \
    ref_substr_1.fasta
```

Some of the reads were sequenced from the reverse complement strand, so they need to be modified before alignment. Example of a fail read  that maps to $[3802876, 3813249]$ in reverse complement form is below.

```python
with open('data/bacillus_subtilis/bs_ref.fasta') as handle:
    for reference in SeqIO.parse(handle, 'fasta'):
        reference.seq = reference.seq[3802876:3813249]
        SeqIO.write(reference, 'ref_substr_2.fasta', 'fasta')

with open('guppy_all/fail/fastq_runid_aea170ff77e7e76bd42c58335c528e15f49ba94c_0_0.fastq') as handle:
    for read in SeqIO.parse(handle, 'fastq'):
        if read.name == '004c4765-e16f-4a98-9cb9-41612ef796e5':
            read.seq = read.seq.reverse_complement()
            SeqIO.write(read, 'read_2.fasta', 'fasta')
            break;
```

```python
!edlib-aligner \
    -p \
    read_2.fasta \
    ref_substr_2.fasta
```

The obtained edit distance gives approximate base accuracy of `0,76177`.

We will use Guppy's high accuracy configuration file to basecall the first `n` Bacillus subtilis reads set and compare the results with the fast mode.

```python
# create a subset
import h5py
n = 5

subset = h5py.File('subset.fast5', mode='w')

dataset = h5py.File('data/bacillus_subtilis/bs_raw.fast5', mode='r')
for read in dataset:
    dataset.copy(read, subset)
    if (len(subset.keys()) == n):
        break

subset.close()
```
