# Stitching fragments

The shortcoming of sequencing technologies is that the read length is considerably smaller than
the sequenced DNA molecule. Reads are often considered as pieces of a jigsaw puzzle which need to be interlocked in order to reconstruct the original picture. This is a task for software called assemblers, which are designed under the assumption that similar reads originate from the same genomic region. This is often invalidated due to sequencing errors and repetitive nature of the genome.

There are two distinct approaches to assembly. One is based on the overlap-layout-consensus paradigm (OLC) describing the relationship between reads, while the other builds a de Bruijn graph from short substrings of the read set. Designed for the first generation of sequencing, the modular OLC paradigm builds an overlap graph from approximate overlaps between each pair of reads. Vertices of the overlap graph encapsulate sequences while the suffix-prefix overlaps between them define the weight and orientation of edges. After applying several graph simplification methods, appropriate paths of the graph are extracted and the sequencing errors are amended with multiple sequence alignment of all reads.

```python
import networkx

# R: GTCAGCCCAGACAGTTTTGCGGATATATAGCGACCGATACGACGTTTATACCCCCC
# O: GTCAGCCCACACAGAT
# 1:             AGTTTTGC
# 2:               TT  GCGGA  TATAGCGAC
# 3:      CCCAGA
# 4:                    CGGATATAT
# 5:                           ATCCCGACGGAAACGA
# 6:                                     ATACGAC TTT
# 7:                                        CGACGTT ATACCCCCC            

reads = ['GTCAGCCCACACAGAT', 'AGTTTTGC', 'TTGCGGATATAGCGAC', 'CCCAGA', 'CGGATATAT', 'ATCCCGACGGAAACGA', 'ATACGACTTT', 'CGACGTTATACCCCCC']

# 0: GTCAGCCCACACAGAT     # 0: GTCAGCCCACACAGAT               # 0: GTCAGCCCACACAGAT
#                || |     #                  ||               #          |||| |    
# 1:             AGTTTTGC # 2:               TTGCGGATATAGCGAC # 3:       CCCAGA    

# 1: AGTTTTGC             # 1: AGTTTTGC        
#      ||  ||             #           |        
# 2:   TT--GCGGATATAGCGAC # 4:        CGGATATAT

# 2: TTGCGGA--TATAGCGAC # 2: TTGCGGATATAGCGAC
#       ||||  |||       #            ||  ||||
# 4:    CGGATATAT       # 5:         ATCCCGACGGAAACGA

# 4: CGGATATAT
#           ||
# 5:        ATCCCGACGGAAACGA

# 5:  ATCCCGACGGAAACGA     # 5: ATCCCGACGGAAACGA            
#               | ||||     #                 |||            
# 6:            ATACGACTTT # 7:              CGACGTTTACCCCCC

# 6: ATACGAC-TTT
#       |||| ||
# 7:    CGACGTT-ATACCCCCC

# Contained = 3 and 4

og = networkx.DiGraph()
og.add_edge(reads[0], reads[1], weight = 12)
og.add_edge(reads[0], reads[2], weight = 14)
og.add_edge(reads[1], reads[2], weight = 2)
og.add_edge(reads[2], reads[5], weight = 8)
og.add_edge(reads[5], reads[6], weight = 10)
og.add_edge(reads[5], reads[7], weight = 13)
og.add_edge(reads[6], reads[7], weight = 3)

print("Nodes =", og.nodes())
```

```python
from matplotlib import pyplot
import seaborn

seaborn.set()
seaborn.set_style("white")

pyplot.rcParams["figure.figsize"] = (20,5)

pos = networkx.shell_layout(og)
networkx.draw(og, pos, node_size = 2500, labels = {n: n for n in og.nodes()})
_ = networkx.draw_networkx_edge_labels(og, pos, edge_labels = dict([((u, v, ), e['weight']) for u, v, e in og.edges(data = True)]))
```

```python
print('Reconstruction: ', end = '')
path = networkx.algorithms.tournament.hamiltonian_path(og)
for i in range(len(path) - 1):
    print(path[i][0:og[path[i]][path[i + 1]]['weight']], end = '')
print(path[-1])
```

The huge increase of sequencing yield in form of short and accurate reads present in second generation of sequencing was problematic for the quadratic complexity of the overlap phase. This led to development of an
algorithm which omits the need for pairwise overlaps. Reads are sliced into overlapping substrings of length $k$, called $k$-mers, and transformed into edges of a de Bruijn graph. Distinct $k-1$ long prefixes and suffixes of all $k$-mers become vertices of the graph. The optimal value for $k$ depends on various factors, such as sequencing depth and the error rate. The assembly problem of finding a Hamiltonian path in overlap graphs is replaced with finding a Eulerian path in de Bruijn graphs, for which a linear algorithm is known.

```python
# GTCAGCCCAGACAGTTTTGCGGATATATAGCGACCGATACGACGTTTATACCCCCC
# GTC   CCA   AGT   GCG   ATA   CGA   ATA   CGT   TAC
#  TCA   CAG   GTT   CGG   TAT   GAC   TAC   GTT   ACC
#   CAG   AGA   TTT   GGA   ATA   ACC   ACG   TTT   CCC
#    AGC   GAC   TTT   GAT   TAG   CCG   CGA   TTA   CCC
#     GCC   ACA   TTG   ATA   AGC   CGA   GAC   TAT   CCC
#      CCC   CAT   TGC   TAT   GCG   GAT   ACG   ATA   CCC

read = 'GTCAGCCCAGACAGTTTTGCGGATATATAGCGACCGATACGACGTTTATACCCCCC'
k = 3

dbg = networkx.MultiDiGraph()

kmers = []
for i in range(len(read) - k + 1):
    kmers.append(read[i:i + k])

print("Number of k-mers =", len(kmers))
print("Number of unique k-mers =", len(set(kmers)))

for kmer in kmers:
    dbg.add_edge(kmer[0:k - 1], kmer[1:k], label = kmer)

print("Nodes =", dbg.nodes())
```

```python
pos = networkx.spiral_layout(dbg)
networkx.draw(dbg, pos, node_size = 2500, labels = {n: n for n in dbg.nodes()})
_ = networkx.draw_networkx_edge_labels(dbg, pos, edge_labels= dict([((u, v, ), e['label']) for u, v, e in dbg.edges(data=True)]))
```

```python
i = 0
print('Reconstruction: ', end = '')
for u, v in networkx.eulerian_path(dbg):
    if i == 0:
        print(u, end = '')
    print(v[k-2], end = '')
    i += 1
```

The possibility of multiple Eulerian paths hinders the discovery of the correct assembly path, so similar simplification methods as in the OLC paradigm are applied. The de Bruijn graph approach is best suited for sequences with high accuracy, but recently researches were successful in generalizing them for third generation of sequencing. On the other hand, the OLC paradigm is tolerant to arbitrary error rates given a sensitive enough overlapper.

9	int	Target end on original strand (0-based)

Minimap2 can also find alignments which we will use to calculate the average accuracy of our data. This can be done by passing parameter `-a` which will print the output in SAM format, but we can also use `-c` to let minimap2 put the CIGAR string in the last column of the PAF file.

```python
!minimap2 \
    -x map-ont \
    -c \
    --eqx \
    -o bs_reads_to_ref.cigar.paf \
    data/bacillus_subtilis/bs_ref.fasta \
    data/bacillus_subtilis/SRR15000906.fastq
```

```python
paf = pandas.read_csv('bs_reads_to_ref.cigar.paf', header = None, delimiter = '\t', error_bad_lines = False)
paf.drop_duplicates(subset = 0, keep = 'first', inplace = True)
paf
```

```python
def base_accuracy(paf):
    stats = {'=': [], 'X': [], 'I': [], 'D': []}
    for i, row in paf.iterrows():
        if pandas.isnull(row[22]):
            continue
        read = {'=': 0., 'X': 0., 'I': 0., 'D': 0.}
        cigar = row[22][5:]
        n = ''
        for c in cigar:
            if c in ['=', 'X', 'I', 'D']:
                read[c] += int(n)
                n = ''
            else:
                n += c
        for m, n in read.items():
            stats[m].append(n / sum(read.values()))

    return stats

stats = base_accuracy(paf)
```

```python
import statistics

def print_statistics(stats):
    print('  Min      Median   Mean     Max')
    for m, n in stats.items():
        print(m, '{0:.6f}'.format(min(n)), '{0:.6f}'.format(statistics.median(n)), '{0:.6f}'.format(statistics.mean(n)), '{0:.6f}'.format(max(n)))

print_statistics(stats)
```
